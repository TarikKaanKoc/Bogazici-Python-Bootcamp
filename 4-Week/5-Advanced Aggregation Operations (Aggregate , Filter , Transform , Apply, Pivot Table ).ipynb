## Advanced Aggregation Operations (Aggregate , Filter , Transform , Apply )
---
Pandas içerisinde bulunan (Bizler için ileri fonksiyonlar) bu fonksiyonlar bizim için hayati öneme sahip, çünkü gerçek hayatta gerçek bir veri ile karşılaştığımızda bu veri üzerinde atlamamız gereken takla olarak nitelendirebiliriz bu fonksiyonları .

---

* aggregate() -> describe' ın aksine tüm aggregation operasyonları yerine kullanıcının istediği aggregation operasyonları için kullanılır.
* Aynı şekilde değişken bazında aggregation işlemi uygulanabilir. (Her değişken için ayrı ayrı aggregation işlemi )
---
>> **Uygulamalı Örnekler ile ele alacağız bu konuyu:**
---

**Ele alınacak olan örnekler:** 

**Örnek1 :** sadece min, max aggregation operasyonunun bir arada kullanmak.

**Örnek2:** sadece min,median,max aggreagtion operasonlarını bir arada kullanmak.

**Örnek3 :** Değişken bazında aggregation işlemi uygulamak. (Her değişken için ayrı ayrı operasyonlar oluşturmak. ) Değişken-1 için min, Değişken-2 için max aggregation.

----

## Filter :
---
**filter() ->** Kendi kişisel fonksiyonlarımızı bir filtreleme aracı olarak değişkenler üzerinde sorgulama işlemi! 

**Örnek:** Değişken-1 üzerinde bulunan grupların standart sapması alınsın. Gözlem birimlerinin standart sapması 9'dan büyük ise bu değerleri görelim. (Tüm gözlem birimleri görüntülenmelidir.) 

---

## Transform & Apply

**transfrom() ->** Veri setimizin değişkenlerine dönüştürme işlemi yapmak istiyoruz ve bu dönüştürme işlemini 
kendimiz tanımlamak istiyoruz. 

**apply() ->** applay fonksiyonu tıpkı transform ve filter fonksiyonu gibi değişkenlerin üzerinde gezinme yeteneği olan ve aggregate (toplulaştırma) kullanılabilecek bir fonksiyondur.

## Aggregate 

```python
import pandas as pd 
import numpy as np

df = pd.DataFrame({'Groups':['A','B','C','A','B','C'],
                  'Variable-1': [10,23,33,22,11,99],
                  'Variable-2':[100,253,333,262,111,969]})
df 
```

```python
df.groupby('Groups')
```

```python
#Örnek1 : sadece min, max aggregation operasyonunun bir arada kullanmak.
df.groupby('Groups').aggregate([min,max])
```

```python
#Örnek2 : sadece min,median,max aggreagtion operasonlarını bir arada kullanmak.
df.groupby('Groups').aggregate([min,np.median,max])
```

---
#### Yukarıdaki kod da np.median yazılmasının nedeni: 

* Median numpy kütüphanesinde yazılan bir fonksiyon , Median pandas mödülünden çekilmediğinden dolayı geldiği yerin adresini belirleyip, nokta eklediğimiz vakit alt bileşenlerine erişebiliyoruz. Bizlerde median yazıp, median fonksiyonuna (bileşenine) eriştik. 
* min , max ise pandas'ta bulunan fonksiyonlar. Direkt yazdığımız vakit fonksiyonları çekip, kullanabiliyoruz. Eğer : Pandasın içerisinde yer alan ve varolduğunu bildiğimiz fonksiyonları kullanmak istersek tırnak içerisinde veya olduğu gibi yazdığımızda direkt erişebiliriz. 
Örnek : min , max -veya- "min","max" <--- Her iki durumda da çalışır. Çünkü bu fonksiyonlar Pandas içeriisnde yer almaktadır.  ( Aggregation fonksiyonu bunu bağlamak üzere algılayabiliyor. )
---

```python
#Örnek3 : Değişken bazında aggregation işlemi uygulamak. (Her değişken için ayrı ayrı operasyonlar oluşturmak. ) 
# Değişken-1 için min, Değişken-2 için max operation.
df.groupby('Groups').aggregate({'Variable-1':min,'Variable-2':max})
```

# Filter 

```python
df = pd.DataFrame({'Groups':['A','B','C','A','B','C'],
                  'Variable-1': [10,23,33,22,11,99],
                  'Variable-2':[100,253,333,262,111,969]})
df 
```

```python
# Burada grupların standart sapması görünmektedir. 
# Ancak mevcut olan tüm gözlem birimlerinin standart sapması görünmüyor ve 9'dan büyük olanlar görütülenmedi.
df.groupby('Groups')['Variable-1'].std()
```

```python
def filter_func(x):
    return x['Variable-1'].std() > 9
```

```python
df.groupby('Groups').filter(filter_func)
```

```python
# Yukarıda gözlemlendiği üzere tüm gözlemlerin standart sapması 9'dan büyük olanlar görüntülendi.
```

## Transform 

* Transform fonksiyonu vektörel çalışan bir fonksiyondur. (Yani sütün bazında, değişken bazında...) işlem yapmak için kullanılan fonksiyon.

```python
df = pd.DataFrame({'Groups':['A','B','C','A','B','C'],
                  'Variable-1': [10,23,33,22,11,99],
                  'Variable-2':[100,253,333,262,111,969]})
df 
```

```python
df_a = df.iloc[:,1:3]
```

```python
df_a.transform(lambda x: x-x.mean())
```

```python
df_a.transform(lambda x: (x-x.mean()) / x.std())
```

----
## Applay 

* applay fonksiyonu tıpkı transform ve filter fonksiyonu gibi değişkenlerin üzerinde gezinme yeteneği olan ve aggregate (toplulaştırma) kullanılabilecek bir fonksiyondur. 
---

```python
df = pd.DataFrame({'Groups':['A','B','C','A','B','C'],
                  'Variable-1': [10,23,33,22,11,99],
                  'Variable-2':[100,253,333,262,111,969]})
df 
```

```python
df.apply(np.sum)
```

# Pivot Tablolar 

* veri setleri üzerinde bazı satır ve sütün işlemlerini gerçekleştirerek veri setini amaca uygun bir biçime dönüştürmek için kullanılan yapılardır. 
---

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#pivot ile table
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class")
```

```python
titanic.age.head()
```

```python
# pivot table ile: 
```

```python
age = pd.cut(titanic["age"], [0, 18, 90])
age.head(10)
```

```python
titanic.pivot_table("survived", ["sex", age], "class")
```
