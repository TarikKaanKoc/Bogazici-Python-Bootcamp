## Join Operations with Pandas (Pandas İle (Join) Birleştirme İşlemleri ) 
---

```python
import pandas as pd
import numpy as np
```

```python
m = np.random.randint(1,30 , size = (10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

```python
df2 = df + 100
```

```python
df2
```

```python
pd.concat([df,df2], ignore_index=True)
```

---
> **Yukarıda gözlemlendiği üzere önceki birleştirme (join) işleminde gözlem birimlerinde bir sorun oluşmuştu , şimdi ise değişken birimlerinde bir sorun görünüyor. Bu durumda yine fonksiyonun işlevseliği, alabileceği argümanları incelemenizi tavsiye ederim!**
---

```python
#Şimdi ise birleştirme (join) argümanı kullanılarak kesişime göre birleştirme işlemi uygulandı 
pd.concat([df,df2], join = "inner" , ignore_index = True )
```

```python
pd.concat([df,df2.reindex(columns = df.columns)] , ignore_index = True )
```
